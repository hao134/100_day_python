## Use BeautifulSoup get the data from the website: https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors

```python
from bs4 import BeautifulSoup
import requests

records = []
for current_page in range(34):
  response = requests.get(f"https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors/page/{current_page + 1}")
  page = response.text
  soup = BeautifulSoup(page, "html.parser")
  rows = soup.select(selector="div table tbody tr")
  for row in rows:
    cells = row.select("span.data-table__value")
    #print([cell.getText().strip() for cell in cells])
    record = {
        "Undergraduate Major": cells[1].getText(),
        "Starting Median Salary": float(cells[3].getText().strip("$").replace(",","")),
        "Mid-Career Median Salary": float(cells[4].getText().strip("$").replace(",",""))
    }
    records.append(record)

pd.DataFrame(records).to_csv("salaries_by_college_major_updated.csv", index = False)
```
